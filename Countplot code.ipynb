```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load Dataset
df = pd.read_csv("Test_data.csv")

print(df.head())
print(df.shape)

# Create a folder to save charts
output_folder = "HR_Charts"
os.makedirs(output_folder, exist_ok=True)

# Categorical Columns
categorical_columns = [
    'department',
    'region',
    'education',
    'gender',
    'recruitment_channel',
    'KPIs_met >80%',
    'awards_won?'
]

# Display Unique Values
for col in categorical_columns:
    print("="*60)
    print(f"{col}")
    print("="*60)
    print(df[col].value_counts())
    print()

# Generate and Save Count Plots
for col in categorical_columns:

    plt.figure(figsize=(10,5))

    df[col].value_counts().plot(
        kind='bar',
        color='skyblue',
        edgecolor='black'
    )

    plt.title(f'Count Plot of {col}', fontsize=14)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()

    # Create a valid filename
    filename = col.replace(" ", "_").replace(">", "gt").replace("?", "")

    # Save chart as PNG
    plt.savefig(
        os.path.join(output_folder, f"{filename}.png"),
        dpi=300,
        bbox_inches='tight'
    )

    plt.show()

    # Close figure to free memory
    plt.close()

print(f"\nAll charts have been saved in the '{output_folder}' folder.")
```
